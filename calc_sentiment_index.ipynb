```python
import pandas as pd
```

```python
import os
```

```python
import numpy as np
```

```python
for file in os.listdir('data/posneg/'):
    filename = file.split('.')[0]
    source = filename.split('_')[1]
    symbol = filename.split('_')[2]
    type = filename.split('_')[3]
    new_file_name = 'pred_' + source + '_' + type + '_si3_' + symbol + '.csv' 
    
    data = pd.read_csv('data/posneg/' + file, index_col=0)
    a = ((data.pos - data.neg) / (data.pos + data.neg + 1))
    a = a + np.abs(a.min()) + 0.01
    print(a.min())
    a = np.log(a.values[1:] / a.values[:1])
    a = np.insert(a, 0, 0)
    
    data['pred'] = a
    data[['pred']].to_csv('data/pred/' + new_file_name)
    
    print(file, len(data))
```
